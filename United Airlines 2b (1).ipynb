```python
dataframes = [
    unaccompanied_minor, checkout, products_and_services, 
    schedule_change, checkin, other_topics, etc, baggage, voluntary_change,nan, 
     mileage_plus, irrops, upgrade, traveler_updates, 
    communications, seating, postflight, voluntary_cancel, disability, 
    digital_support, booking
]
reason_dict = {}

for i, df in enumerate(dataframes, 1):
    key_name = f'reason_{i}'  # Create keys like reason_1, reason_2, etc.
    reason_dict[key_name] = df  # Assign the DataFrame to the key

# Optional: Print the dictionary keys to confirm
print(reason_dict.keys())
```

```python
for reason,df in reason_dict.items():
    if 'call_start_datetime' in df.columns:
        df['hour'] = pd.to_datetime(df['call_start_datetime']).dt.hour
    
```

```python

selected_columns = ['call_id','call_start_datetime','agent_assigned_datetime','call_end_datetime','primary_call_reason','dominant_topic','topic_keywords','hour']  # Replace with actual column names

filtered_reason_dict = {}

for key, df in reason_dict.items():
    available_columns = [col for col in selected_columns if col in df.columns]
    
    filtered_reason_dict[key] = df[available_columns]
    
    print(f"Filtered DataFrame for {key}:")
    print(filtered_reason_dict[key].head())
```

```python
grouped_reason_dict = {}

for key, df in filtered_reason_dict.items():
    if 'dominant_topic' in df.columns: 
        grouped_reason_dict[key] = df.groupby('dominant_topic').size()  # Replace `.sum()` with desired aggregation function
    
        # Optional: Print to check the groupby operation
        print(f"Grouped DataFrame for {key}:")
        print(grouped_reason_dict[key].head())

# Now grouped_reason_dict holds the grouped DataFrames
```

```python
for reason, df in filtered_reason_dict.items():
    if all(col in df.columns for col in ['call_start_datetime', 'agent_assigned_datetime', 'call_end_datetime']):
        df['handle_time'] = (pd.to_datetime(df['call_end_datetime']) - pd.to_datetime(df['agent_assigned_datetime'])).dt.total_seconds() / 60
        df['wait_time'] = (pd.to_datetime(df['agent_assigned_datetime']) - pd.to_datetime(df['call_start_datetime'])).dt.total_seconds() / 60
```

```python
handle_time_by_topic = {}
for reason,df in filtered_reason_dict.items():
    if all(col in df.columns for col in ['handle_time', 'dominant_topic', 'primary_call_reason']):
        grouped_df = pd.DataFrame(df.groupby(['dominant_topic'])['handle_time'].mean())
        
        handle_time_by_topic[reason] = {
                'grouped_handle_time': grouped_df,
                'primary_call_reason': df['primary_call_reason'].unique()  # Store unique values or all rows if needed
            }
    
```

```python
grouped_reason_dict = {}

for key, df in filtered_reason_dict.items():
    if 'dominant_topic' in df.columns: 
        # Group by 'dominant_topic' and count occurrences
        grouped_df = df.groupby('dominant_topic').size().reset_index(name='count')
        
        # Add the primary call reason
        primary_reason = df['primary_call_reason'].unique()[0]  # Get unique primary call reason
        grouped_df['primary_call_reason'] = primary_reason
        
        # Store the DataFrame in the dictionary
        grouped_reason_dict[key] = grouped_df

        # Optional: Print to check the groupby operation
        print(f"Grouped DataFrame for {key}:")
        print(grouped_reason_dict[key].head())
```

```python
grouped_reason_dict['reason_1']
```

```python
combined_df = pd.concat(grouped_reason_dict.values(), ignore_index=True)
combined_df
```

```python
import pandas as pd

# Initialize an empty list to hold the rows for the final DataFrame
rows = []

# Iterate over the handle_time_by_topic dictionary
for reason, data in handle_time_by_topic.items():
    # Extract the grouped handle time DataFrame
    grouped_handle_time = data['grouped_handle_time']
    # Extract the unique primary call reason for the current reason
    primary_call_reason = data['primary_call_reason'][0]  # Assuming there's at least one unique value

    # Iterate over the grouped handle time DataFrame to extract handle time and dominant topics
    for dominant_topic, row in grouped_handle_time.iterrows():
        # Append a new row to the rows list
        rows.append({
            'primary_call_reason': primary_call_reason,
            'handle_time': row['handle_time'],
            'dominant_topic': dominant_topic
        })

# Create a DataFrame from the list of rows
final_handle_time_df = pd.DataFrame(rows)

# Display the resulting DataFrame
print(final_handle_time_df)
```

```python
final_df = final_handle_time_df.merge(combined_df,on=['primary_call_reason','dominant_topic'])
```
